```python
# Verificando se o Viewership Score possui outliers
df_netflix_daily_top_10.plot.box(column='Viewership Score')
```

```python
#  Histograma - Checar visualmente se os dados possuem uma distribuição normal
df_netflix_daily_top_10['Viewership Score'].plot.hist()
```

```python
# Método de Tukey - IQR (Distribuição não normal)
# IQR - Range Interquartil
# IQR = terceiro quartil - primeiro quartil

q1_netflix_viewership_score = df_netflix_daily_top_10['Viewership Score'].quantile(0.25)
q3_netflix_viewership_score = df_netflix_daily_top_10['Viewership Score'].quantile(0.75)

iqr_netflix_viewership_score = q3_netflix_viewership_score - q1_netflix_viewership_score
```

```python
iqr_netflix_viewership_score
```

```python
# Limites Inferior e Superior
limInf_netflix_viewership_score = q1_netflix_viewership_score - (iqr_netflix_viewership_score * 1.5)
limSup_netflix_viewership_score = q3_netflix_viewership_score + (iqr_netflix_viewership_score * 1.5)
limInf_netflix_viewership_score, limSup_netflix_viewership_score
```

```python
df_netflix_daily_top_10[(df_netflix_daily_top_10['Viewership Score'] < limInf_netflix_viewership_score) | (df_netflix_daily_top_10['Viewership Score'] > limSup_netflix_viewership_score)]
```
